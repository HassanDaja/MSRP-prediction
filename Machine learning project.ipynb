Dealing With Nan values

```python
for col in data.columns:
  print(f'{col}:{(data[col].isnull().sum()/data.shape[0])*100:.3f}%')
```

```python
#Taking look at the 3 recoreds with nan value in
data[data['Engine Fuel Type'].isna()]
```

```python
#replacing the nan recordes using the Mode of each Company fuel
import numpy as np
Fuel_type=data.groupby(['Make'])['Engine Fuel Type'].agg(pd.Series.mode)
def replace_nan(value):
    if pd.isnull(value['Engine Fuel Type']):
      return Fuel_type[value['Make']]
    else:
      return value['Engine Fuel Type']
```

```python
data['Engine Fuel Type']=data.apply(lambda x:replace_nan(x),axis=1)
```

```python
#splitting cat and num data so we can work on each one alone
cat_data=data.select_dtypes(include='object')
num_data=data.select_dtypes(exclude='object')
```

```python
#Getting the catigorical columns with 16 uniuque data or less
for col in cat_data.columns:
  if 16<len(cat_data[col].unique()):
    cat_data.drop(col,axis=1,inplace=True)
```

```python
cat_data
```

```python
cat_data=pd.get_dummies(cat_data)
```

```python
num_data['Year']=num_data['Year'].apply(lambda x:2022-x)
```

```python
#Normalizing num data
from sklearn import preprocessing
num_data = pd.DataFrame(preprocessing.normalize(num_data),columns=num_data.columns)
```

```python
#df JOIN
def make_df(num,cat):
  df=pd.DataFrame()
  for col in num:
    df[col]=num[col]
  for col in cat:
    df[col]=cat[col]
  return df
```

```python
new_df=make_df(num_data,cat_data)
```

```python
#Ranking every car company depends on the avg MSRP
new_df['Make']=data["Make"]
avg_price = new_df.groupby('Make')['MSRP'].agg(np.mean).sort_values(ascending=False)
car_type=[]
car_msrp=[]
full_info=[]
for x in avg_price.index:
  car_type.append(x)
for x in avg_price:
  car_msrp.append(x)
for x,y in zip(car_msrp,car_type):
  full_info.append([x,y])
full_info.sort(reverse=True)
dic={}
for pos,val in enumerate(full_info[::-1]):
  dic[val[1]]=pos
new_df['Make']=new_df['Make'].apply(lambda x:dic[x])
```

```python
new_df
```

```python
#Splitting data
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
new_df=new_df.drop('MSRP',axis=1)
y=data['MSRP']
x_train,x_test,y_train,y_test=train_test_split(new_df,y,test_size=.15,random_state=0)
```

```python
#Getting the best number of trees
from sklearn.tree import DecisionTreeRegressor
mae=[]
for leaf in range(2,1500):
  model=DecisionTreeRegressor(max_leaf_nodes=leaf)
  model.fit(x_train,y_train)
  mae.append(mean_absolute_error(y_test,model.predict(x_test)))
```

```python
#Randomforest model
model=RandomForestRegressor(max_leaf_nodes=mae.index(min(mae))+2,n_estimators=1000)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(mean_absolute_error(y_test,y_pred))
```

```python
print(mae.index(min(mae))+2)
```

```python
from sklearn.model_selection import cross_val_score
scoring = 'neg_mean_absolute_error'
mae = cross_val_score(model,new_df, y, cv=10,scoring=scoring)
```
